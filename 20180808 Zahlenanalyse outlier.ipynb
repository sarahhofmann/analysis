```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data/zip.train", header=None, sep=" ")
```

```python
cleaned_data = data.iloc[:,:-1]
```

```python
cleaned_data
```

```python
number9 = pd.DataFrame(cleaned_data[cleaned_data[0] == 9].iloc[:,1:])
data_mean_9 = number9.mean()
data_std_9 = number9.std()
upper9 = data_mean_9.add(2*data_std_9, fill_value=0)
upper9 = upper9.to_frame().transpose()
number_examples__9 = number9.shape[0] #Anzahl der Reihen
upper_number_9 = upper9.values.repeat(number_examples__9) #Matrix
upper_number_9 = np.reshape(upper_number_9, (number_examples__9, len(number9.columns)))
number9_cleaned= number9[number9 < upper_number_9]
number9_nas= np.mean(number9_cleaned.isna().sum(axis=0))
number9_nasstd = np.std(number9_cleaned.isna().sum(axis=0))
number9_superclean = number9_cleaned.dropna(thresh=(number9_nas-2*number9_nasstd), axis=0)
image_mean = np.mean(number9_superclean)
image_reshaped = image_mean.values.reshape((16,16))
plt.imshow(image_reshaped, cmap="magma", interpolation="none")
plt.show()
```

```python
def outlier_function(n):
    number9 = pd.DataFrame(cleaned_data[cleaned_data[0] == n].iloc[:,1:])
    data_mean_9 = number9.mean()
    data_std_9 = number9.std()
    upper9 = data_mean_9.add(2*data_std_9, fill_value=0)
    upper9 = upper9.to_frame().transpose()
    number_examples__9 = number9.shape[0] #Anzahl der Reihen
    upper_number_9 = upper9.values.repeat(number_examples__9) #Matrix
    upper_number_9 = np.reshape(upper_number_9, (number_examples__9, len(number9.columns)))
    number9_cleaned= number9[number9 < upper_number_9]
    number9_nas= np.mean(number9_cleaned.isna().sum(axis=0))
    number9_nasstd = np.std(number9_cleaned.isna().sum(axis=0))
    number9_superclean = number9_cleaned.dropna(thresh=(number9_nas-2*number9_nasstd), axis=0)
    image_mean = np.mean(number9_superclean)
    image_reshaped = image_mean.values.reshape((16,16))
    plt.imshow(image_reshaped, cmap="magma", interpolation="none")
    return plt.show()

outlier_function(0)
outlier_function(1)
outlier_function(2)
outlier_function(3)
outlier_function(4)
outlier_function(5)
outlier_function(6)
outlier_function(7)
outlier_function(8)
outlier_function(9)

    
    
    
```
